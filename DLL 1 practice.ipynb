```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def activationFunction(value,thresholdValue):
  if(value>=thresholdValue):
    return 1
  else:  
    return 0
```

```python
inputVector=np.random.choice([0,1],3)
weightVector=np.random.choice([-1,1],3)
print("Input vector: ", inputVector)
print("Weight vector: ", weightVector)
```

##Mp Neuron

```python
thresholdValue=1
mpSum=np.sum(inputVector)
mpActivation=activationFunction(mpSum,thresholdValue)
print("mpActivation",mpActivation)
```

#Perceptron

```python
dotProduct=inputVector@weightVector;
print(dotProduct)
```

```python
perceptronActivation=activationFunction(dotProduct,thresholdValue)
print("perceptron Activation",perceptronActivation)
```

# Sigmoid Neuron

```python
def sigmoidActivationFunction(x):
  return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10,30)
y=sigmoidActivationFunction(x)
plt.ylabel("Sigmoid Function")
plt.xlabel("x")
plt.plot(x, y) 
plt.show()
```
